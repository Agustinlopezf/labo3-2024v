```python
ventas_totales = sell_in.groupby(['periodo'])['tn'].sum()
ventas_totales = ventas_totales.reset_index()
ventas_totales['Timestamp'] = pd.to_datetime(ventas_totales['periodo'], format='%Y%m')
ventas_totales.drop(columns=['periodo'], inplace=True)


ventas_totales
```

```python
import matplotlib.pyplot as plt

# Ventas totales por mes
plt.plot(ventas_totales['Timestamp'], ventas_totales['tn'], marker = 'o')
plt.xlabel('Periodo')
plt.ylabel('Venta en tn')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import os
import numpy as np
from sklearn.cluster import KMeans
from fastdtw import fastdtw
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt



nombre_experimento = 'ensamble_predicciones'
nombre_archivo_ensamble = 'ensamble_predicciones.csv'


carpeta_datasets = 'C:\\Users\\alope\\Desktop\\Trámites\\Maestria Data Science - Universidad Austral\\Laboratorio de implementación 3\\Datos'
carpeta_exp_base = 'C:\\Users\\alope\\Desktop\\Trámites\\Maestria Data Science - Universidad Austral\\Laboratorio de implementación 3\\Resultados'


carpeta_exp = os.path.join(carpeta_exp_base, nombre_experimento)
if not os.path.exists(carpeta_exp):
    os.makedirs(carpeta_exp)
    


   
```

```python
# Listar todos los archivos en la carpeta y filtrar solo los .csv
archivos_csv = [archivo for archivo in os.listdir(carpeta_exp) if archivo.endswith('.csv') and archivo != nombre_archivo_ensamble]

df_completo = pd.DataFrame()
# Imprimir la lista de archivos CSV
for archivo in archivos_csv:
    df = pd.read_csv(os.path.join(carpeta_exp, archivo))
    df['Experimento'] = archivo
    df_completo = pd.concat([df_completo, df], axis = 0)
    
df_completo
```

```python
df_ensamble = df_completo.groupby(['product_id'])['tn'].mean().reset_index()
df_ensamble
```

```python
df_ensamble.to_csv(os.path.join(carpeta_exp, 'ensamble_predicciones.csv'), index = False)
```
